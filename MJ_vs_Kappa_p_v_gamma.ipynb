```python
%matplotlib inline
import numpy as np
import pylab as pl
import scipy.special as special
from scipy.integrate import quad
```

```python
#want to determine/test the functional form of kappa in terms of gamma
m = 1
c = 1 

p = np.arange(-10, 10, 0.1)

gamma = np.sqrt(1 + (p/(m*c))**2.)


def kappa_gamma_unnormalized(gamma, theta, kappa):
    p = np.sqrt(m**2. * c**2. * (gamma**2. - 1.))
    f = (1. + (np.sqrt(1. + p**2.) - 1.)/(kappa * theta))**(-kappa - 1.)
    return f

def kappa_gamma_to_be_normalized(gamma, theta, kappa):
    p = np.sqrt(m**2. * c**2. * (gamma**2. - 1.))
    dp = (gamma * m * c) / np.sqrt(gamma**2. - 1)
    f = 4 * np.pi * p**2. * dp * kappa_gamma_unnormalized(gamma, theta, kappa)
    return f

def kappa_gamma_normalized(gamma, theta, kappa):
    norm = quad(lambda gamma: kappa_gamma_to_be_normalized(gamma, theta, kappa), 1, np.inf)
    f = 1./norm[0] * kappa_gamma_unnormalized(gamma, theta, kappa)
    return f

def MJ_gamma_unnormalized(gamma, theta):
    p = np.sqrt(m**2. * c**2. * (gamma**2. - 1.))
    f = np.exp(-gamma/theta)
    return f

def MJ_gamma_to_be_normalized(gamma, theta):
    p = m * c * np.sqrt(gamma**2. - 1)
    dp = (gamma * m * c)/np.sqrt(gamma**2. - 1.)
    f = 4 * np.pi * p**2. * MJ_gamma_unnormalized(gamma, theta) * dp
    return f

def MJ_gamma_normalized(gamma, theta):
    norm = quad(lambda gamma: MJ_gamma_to_be_normalized(gamma, theta), 1, np.inf)
    f = (1./norm[0]) * MJ_gamma_unnormalized(gamma, theta)
    return f
```

```python
#kappa and MJ seem to match on a log-log plot, but not on other plots...?

pl.plot(p, MJ_gamma_normalized(gamma, 10), label='MJ')
pl.plot(p, kappa_gamma_normalized(gamma, 10, 150), label='kappa')
pl.legend()
```

```python
#rewriting MJ and Kappa in terms of velocity, v

#p = np.arange(-10, 10, 0.1)

#v = np.sqrt((-p**2.)/(m**2. * (1 + (p/(m*c))**2.)))

v = np.arange(0, 1, 0.01)

def kappa_v_unnormalized(v, theta, kappa):
    gamma = 1./np.sqrt(1. - v**2./c**2.)
    p = m * c * np.sqrt(gamma**2. - 1.)
    f = (1 + (np.sqrt(1 + p**2.) - 1.)/(kappa * theta))**(-kappa - 1.)
    return f

def kappa_v_to_be_normalized(v, theta, kappa):
    gamma = 1./np.sqrt(1. - v**2./c**2.)
    p = m * c * np.sqrt(gamma**2. - 1.)
    dp = gamma * (gamma**2. - 1.)/(m * c**2.)
    f = 4 * np.pi * p**2. * kappa_v_unnormalized(v, theta, kappa) * dp
    return f

def kappa_v_normalized(v, theta, kappa):
    norm = quad(lambda v: kappa_v_to_be_normalized(v, theta, kappa), 0, 1.)
    f = 1./norm[0] * kappa_v_unnormalized(v, theta, kappa)
    return f

def MJ_v_unnormalized(v, theta):
    gamma = 1./np.sqrt(1. - v**2./c**2.)
    #f = gamma * np.sqrt(gamma**2. - 1.) * np.exp(-gamma / theta)
    f = np.exp(-gamma/theta)
    return f

def MJ_v_to_be_normalized(v, theta):
    gamma = 1./np.sqrt(1. - v**2./c**2.)
    p = m * c * np.sqrt(gamma**2. - 1.)
    dp = gamma * (gamma**2. - 1.)/(m * c**2.)
    f = 4. * np.pi * p**2. * MJ_v_unnormalized(v, theta) * dp
    return f

def MJ_v_normalized(v, theta):
    norm = quad(lambda v: MJ_v_to_be_normalized(v, theta), 0, 1.)
    f = 1./norm[0] * MJ_v_unnormalized(v, theta)
    return f
```

```python
pl.plot(v, MJ_v_normalized(v, 10), label='MJ')
pl.plot(v, kappa_v_normalized(v, 10, 150), label='kappa=150')

pl.legend()
```

```python
#kappa and MJ in terms of momentum, p
m = 1
c = 1

p = np.arange(-10, 10, 0.1)

def kappa_p_unnormalized(p, theta, kappa):
    f = (1 + (np.sqrt(1 + p**2.) - 1.)/(kappa * theta))**(-kappa - 1.)
    return f

def kappa_p_to_be_normalized(p, theta, kappa):
    f = 4 * np.pi * p**2. * kappa_p_unnormalized(p, theta, kappa)
    return f

def kappa_p_normalized(p, theta, kappa):
    norm = quad(lambda p: kappa_p_to_be_normalized(p, theta, kappa), 0, np.inf)
    f = 1./norm[0] * kappa_p_unnormalized(p, theta, kappa)
    return f

def MJ_p_unnormalized(p, theta):
    f = np.exp(-np.sqrt(1 + (p/(m*c))**2.) / theta)
    return f

def MJ_p_to_be_normalized(p, theta):
    f = 4. * np.pi * p**2. * MJ_p_unnormalized(p, theta)
    return f

def MJ_p_normalized(p, theta):
    norm = quad(lambda p: MJ_p_to_be_normalized(p, theta), 0, np.inf)
    f = 1./norm[0] * MJ_p_unnormalized(p, theta)
    return f
```

```python
pl.plot(p, kappa_p_normalized(p, 10, 150), label='kappa')
pl.plot(p, MJ_p_normalized(p, 10), label='MJ')
pl.legend()
```

```python
#checking the functional forms of MJ

p = np.arange(-10, 10, 0.1)

def MJ_gamma_of_p_unnormalized(p, theta):
    gamma = np.sqrt(1 + p**2./(m**2. * c**2.))
    f = np.exp(-gamma / theta)
    return f

def MJ_gamma_of_p_to_be_normalized(p, theta):
    f = 4. * np.pi * p**2. * MJ_gamma_of_p_unnormalized(p, theta)
    return f

def MJ_gamma_of_p_normalized(p, theta):
    norm = quad(lambda p: MJ_gamma_of_p_to_be_normalized(p, theta), 0, np.inf)
    f = 1./norm[0] * MJ_gamma_of_p_unnormalized(p, theta)
    return f

def kappa_gamma_of_p_unnormalized(p, theta, kappa):
    gamma = np.sqrt(1 + p**2./(m**2. * c**2.))
    f = (1. + (np.sqrt(1. + p**2.) - 1.)/(kappa * theta))**(-kappa - 1.)
    return f

def kappa_gamma_of_p_to_be_normalized(p, theta, kappa):
    f = 4. * np.pi * p**2. * kappa_gamma_of_p_unnormalized(p, theta, kappa)
    return f

def kappa_gamma_of_p_normalized(p, theta, kappa):
    norm = quad(lambda p: kappa_gamma_of_p_to_be_normalized(p, theta, kappa), 0, np.inf)
    f = 1./norm[0] * kappa_gamma_of_p_unnormalized(p, theta, kappa)
    return f


pl.plot(p, MJ_gamma_of_p_normalized(p, 10), label='MJ($\\gamma$)')
pl.plot(p, MJ_p_normalized(p, 10), label='MJ(p)')

pl.plot(p, kappa_gamma_of_p_normalized(p, 10, 150), label='$\\kappa$($\\gamma$)')
pl.plot(p, kappa_p_normalized(p, 10, 150), label='$\\kappa$(p)')

pl.legend()
```
