```python
#kappa=3.5 high and low freq emiss
data = np.loadtxt('/home/alex/Documents/paper_plots/kappa_3_5.txt')
x = data[:, 0]
y = data[:, 1]

def kappa_low(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_w
    norm = 4. * np.pi * special.gamma(kappa-4./3.) / (3.**(7./3.) * special.gamma(kappa-2.))
    #norm = C_fit(kappa, 2.43091789,  0.45470688, -3.11539714)
    #norm = C_fit(kappa, 2.43,  0.454, -3.12)
    dimensional = (n_e * e**2. * nu_c * np.sin(theta))/(c)
    ans = norm * dimensional * X**(1./3.)
    return ans

def kappa_high(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_w
    term1 = (1./4.) * 3.**((kappa-1.)/2.) * (kappa-2.) * (kappa-1.)
    term2 = special.gamma(kappa/4. - 1./3.) * special.gamma(kappa/4. + 4./3.) 
    D = term1 * term2
    #D = 10.**D_fit(kappa, 0.5141287 , -0.99470623)
    #D = 10.**D_fit(kappa, 0.5 , -1)
    term3 = (n_e * e**2. * nu_c * np.sin(theta)) / c
    ans = D*term3 * X**(-(kappa-2.)/2.)
    return ans

pl.loglog(x, y, label='symphony')
pl.loglog(x, kappa_low(x), label='$\\bar{j}_{\\nu, low}$', ls='-.', lw='8')
pl.loglog(x, kappa_high(x), label='$\\bar{j}_{\\nu, high}$', ls='--', lw='4')
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("$j_\\nu$")
pl.legend(loc='lower left', prop={'size':35})
pl.show()

pl.semilogx(x, abs(kappa_low(x)/y-1.), label='$\\bar{j}_{\\nu, low}$', ls='-.', lw='8', color='g')
pl.semilogx(x, abs(kappa_high(x)/y-1.), label='$\\bar{j}_{\\nu, high}$', ls='--', lw='4', color='r')
pl.ylim([-1., 1.])
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("$|\\bar{j_\\nu} /j_\\nu - 1|$")
pl.legend(loc='lower left', prop={'size':35})
pl.show()
```
