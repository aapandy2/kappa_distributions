```python
#constants in natural units
m = 1.
c = 1.
theta_e = 10.
e = 1
B = 10
nu_c = (e * B)/(2 * np.pi * m * c)


#constants using iharmony's units
#m = 9.1093826e-28
#c = 2.99792458e10
#theta_e = 10.
#e = 4.80320680e-10
#B = 100.
#n_e = 1.e5
#nu_c = (e * B)/(2 * np.pi * m * c)


#parameters for the kappa distribution
kappa = 3.5
gamma_cutoff = 1000.

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1.e7 * nu_c
theta = np.pi/3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

#max number of terms in sum
n_max = 30

#eq. 67 gives location of peak for n integration for thermal; in
#the n integration we integrate from end of sum (n_max) to C * n_peak_67
C = 1.
n_peak = 1.2e9
        
def K_s(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1 - cos_xi**2.)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1 - cos_xi**2.))/nu_c
    K_xx = M**2. * (special.jv(n, z))**2.
    K_yy = N**2. * (special.jvp(n, z, 1))**2.
    ans = K_xx + K_yy
    return ans    

def kappa_main(gamma):
    cutoff = np.exp(-gamma/gamma_cutoff)
    kappa_body = (1. + (gamma - 1.)/(kappa * theta_e))**(- kappa - 1.)
    return (kappa_body * cutoff)

def norm_term(gamma):
    norm_t = 4. * np.pi * m**3. * c **3. * gamma * np.sqrt(gamma**2 - 1.) * kappa_main(gamma)
    return norm_t

norm_kappa_value = quad(norm_term, 1., np.inf, epsabs = 0., epsrel = 1e-8)[0]
#print quad(norm_term, 1., np.inf)[0] * norm_kappa_value

def kappa_f(gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    f = (1./norm_kappa_value) * kappa_main(gamma)
    return f

def I(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    #this involves f; this current version uses kappa_f; 2pi comes from integrating out dphi
    ans = (2. * np.pi * e**2. * nu**2.)/c * (m**3. * c**3. * gamma**2. * beta * 2 * np.pi) * kappa_f(gamma) * K_s(gamma, n)  
    return ans

def gamma_integrand(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    ans = 1./(nu * beta * abs(np.cos(theta))) * I(gamma, n)
    return ans

def gamma_integration_result(n):
    gamma_minus = ((n*nu_c)/nu - abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    gamma_plus = ((n*nu_c)/nu + abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    
    def integrand(gamma):
        f = gamma_integrand(gamma, n)
        return f
    result = quad(integrand, gamma_minus, gamma_plus, epsabs = 0, epsrel = 1e-08)[0]
    return result

#integrate over n after evaluating (n_max) terms of sum
def n_integration(n_max, n_minus):
    if(n_max < n_minus):
        n_max = n_minus
    ans = quad(gamma_integration_result, n_max, C * n_peak, epsabs = 1e-29, epsrel = 1e-08)[0]
    #print ans
    return ans

#evaluate (n_max) number of terms of the sum
def n_summation(n_max):
    j_nu = 0
    n_minus = (nu/nu_c) * abs(np.sin(theta))
        
    for x in range (int(n_minus+1), int(n_minus + 1 + n_max)):
        j_nu = j_nu + gamma_integration_result(x)
        #print j_nu
    
    j_nu = j_nu + n_integration(n_max, n_minus)
    return j_nu
    

n_summation(n_max)
```
