```python
#constants in natural units
m = 1.
c = 1.
theta_e = 10.
e = 1
B = 10
nu_c = (e * B)/(2 * np.pi * m * c)

#parameters for the power law distribution
p = 3.
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 9. * nu_c
theta = np.pi/3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

#max number of terms in sum
n_max = 30

#eq. 67 gives location of peak for n integration for thermal; in
#the n integration we integrate from end of sum (n_max) to C * n_peak_67
C = 10
n_peak_power_law = 35
        
def K_s(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1 - cos_xi**2.)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1 - cos_xi**2.))/nu_c
    K_xx = M**2. * (special.jv(n, z))**2.
    K_yy = N**2. * (special.jvp(n, z, 1))**2.
    #K_xx = M**2. * (my_Bessel_J(n, z))**2.
    #K_yy = N**2. * (my_Bessel_dJ(n, z, 1))**2.
    ans = K_xx + K_yy
    return ans    

def power_law_f(gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    #d is the differential dn_e/dgammadphid(cos(xi))
    d = (n_e_NT * (p - 1.) * gamma**(-p)) / (4 * np.pi * (gamma_min**(1. - p) - gamma_max**(1. - p)))
    f = 1./(m**3. * c**3. * gamma**2. * beta) * d
    return f
    
def I(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    #this involves f; this current version uses MJ_f; 2pi comes from integrating out dphi
    ans = (2. * np.pi * e**2. * nu**2.)/c * (m**3. * c**3. * gamma**2. * beta * 2 * np.pi) * power_law_f(gamma) * K_s(gamma, n)  
    return ans

def gamma_integrand(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    ans = 1./(nu * beta * abs(np.cos(theta))) * I(gamma, n)
    return ans

def gamma_integration_result(n):
    gamma_minus = ((n*nu_c)/nu - abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    gamma_plus = ((n*nu_c)/nu + abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    
    def integrand(gamma):
        f = gamma_integrand(gamma, n)
        return f
    result = quad(integrand, gamma_minus, gamma_plus, epsabs = 1e-29)[0]
    return result

#integrate over n after evaluating (n_max) terms of sum
def n_integration(n_max, n_minus):
    if(n_max < n_minus):
        n_max = n_minus
    ans = quad(gamma_integration_result, n_max, C * n_peak_power_law, epsabs = 1e-29)[0]
    #print ans
    return ans

#evaluate (n_max) number of terms of the sum
def n_summation(n_max):
    j_nu = 0
    n_minus = (nu/nu_c) * abs(np.sin(theta))
        
    for x in range (int(n_minus+1), int(n_minus + 1 + n_max)):
        j_nu = j_nu + gamma_integration_result(x)
        #print x
        #print j_nu
    
    j_nu = j_nu + n_integration(n_max, n_minus)
    return j_nu

print nu / nu_c
n_summation(n_max)

#print gamma_integration_result(10)
```

```python
#plotting power law j_nu with high and low error tolerance on scipy quad integrator
#plot shown (not commented out) has low error tolerance (epsabs = 1e-12)

#nu_ratio = nu/nu_c
nu_ratio = np.arange(1., 1e6, 1.)

def analytic_soln_eq_58(nu_ratio):
    prefactor = n_e_NT*(e**2.*nu_c)/c*(3.**(p/2.)*(p-1.)*np.sin(theta))/(2.*(p+1.)*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3. * p - 1.)/12.) * special.gamma((3. * p + 19.)/12.)
    ans = prefactor * gamma_fns * (nu_ratio/(np.sin(theta)))**(-(p - 1.)/2.)
    return ans

data_low = np.loadtxt('/home/alex/Documents/jnu_power_law_low_tolerance.txt')
#data_high = np.loadtxt('/home/alex/Documents/jnu_power_law_high_tolerance.txt')
pl.xlabel("$\\log_{10}(\\nu$/$\\nu_c$)")
pl.ylabel("log(j$_\\nu$)")



pl.loglog(data_low[:, 0], data_low[:, 1], label='j$_\\nu$ epsabs=1e-12')
#pl.loglog(data_high[:, 0], data_high[:, 1], label='j$_\\nu$ epsabs=1e-2')
pl.loglog(nu_ratio, analytic_soln_eq_58(nu_ratio), label='Eq. 58')
pl.legend()
pl.show()
```
