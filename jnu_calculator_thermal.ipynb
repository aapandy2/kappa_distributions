```python
#want to evaluate jnu using some integration tricks
#to make the integral doable

#constants in natural units
m = 1.
c = 1.
theta_e = 10.
e = 1
B = 10
nu_c = (e * B)/(2 * np.pi * m * c)

#parameters for the power law distribution
p = 3.
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1.e5 * nu_c
theta = np.pi/3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

#max number of terms in sum
n_max = 30

#eq. 67 gives location of peak for n integration for thermal; in
#the n integration we integrate from end of sum (n_max) to C * n_peak_67
C = 10
def n_peak(nu):
    if(nu <= nu_c * theta_e**2.):
        beta_67 = np.sqrt(1. - 1./(1. + theta_e)**2.)
        #n_peak_67 = (theta_e + 1.) * (nu/nu_c) * (1 - beta_67**2.*(np.cos(theta))**2.)
        #n_peak = 1./np.sqrt(1. - beta_67**2.) * (nu/nu_c) * (1. - beta_67**2. * (np.cos(theta))**2.)
        n_peak = (theta_e + 1. + (2. * theta_e * nu / nu_c)**(1./3.)) * (nu/nu_c) * (1 - beta_67**2. * (np.cos(theta))**2.)
        return n_peak
    elif(nu > nu_c * theta_e**2.):
        beta_68 = np.sqrt(1. - (2. * theta_e * nu / nu_c)**(-2./3.))
        #n_peak = 1./np.sqrt(1 - beta_68**2.) * (nu/nu_c) * (1. - beta_68**2. * (np.cos(theta))**2.)
        n_peak = (theta_e + 1. + (2. * theta_e * nu / nu_c)**(1./3.)) * (nu/nu_c) * (1 - beta_68**2. * (np.cos(theta))**2.)
        return n_peak

    
#def bessel_der(n, z):
#    x = sympy.var('x')
#    diff_bessel = sympy.lambdify(x, sympy.diff(sympy.besselj(n, x), x))
#    return np.float64(diff_bessel(z))
    
def K_s(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1 - cos_xi**2.)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1 - cos_xi**2.))/nu_c
    K_xx = M**2. * (special.jv(n, z))**2.
    K_yy = N**2. * (special.jvp(n, z, 1))**2.
    ans = K_xx + K_yy
    return ans

def power_law_f(gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    #d is the differential dn_e/dgammadphid(cos(xi))
    d = (n_e_NT * (p - 1.) * gamma**(-p)) / (4 * np.pi * (gamma_min**(1. - p) - gamma_max**(1. - p)))
    f = 1./(m**3. * c**3. * gamma**2. * beta) * d
    return f
    
def MJ_f(gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    #d is the differential dn_e/dgammadphid(cos(xi))
    d = (n_e * gamma * np.sqrt(gamma**2. - 1.) * np.exp(-gamma/theta_e))/(4 * np.pi * theta_e * special.kn(2., 1./theta_e))
    f = 1./(m**3. * c**3. * gamma**2. * beta) * d
    return f

def I(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    #this involves f; this current version uses MJ_f; 2pi comes from integrating out dphi
    ans = (2. * np.pi * e**2. * nu**2.)/c * (m**3. * c**3. * gamma**2. * beta * 2 * np.pi) * MJ_f(gamma) * K_s(gamma, n)  
    return ans

def gamma_integrand(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    ans = 1./(nu * beta * abs(np.cos(theta))) * I(gamma, n)
    return ans

def gamma_integration_result(n):
    gamma_minus = ((n*nu_c)/nu - abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    gamma_plus = ((n*nu_c)/nu + abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    
    def integrand(gamma):
        f = gamma_integrand(gamma, n)
        return f
    result = quad(integrand, gamma_minus, gamma_plus)[0]
    return result

#integrate over n after evaluating (n_max) terms of sum
def n_integration(n_max, n_minus):
    if(n_max < n_minus):
        n_max = n_minus
    ans = quad(gamma_integration_result, n_max, C * n_peak(nu))[0]
    #print ans
    return ans

#evaluate (n_max) number of terms of the sum
def n_summation(n_max):
    j_nu = 0
    n_minus = (nu/nu_c) * abs(np.sin(theta))
        
    for x in range (int(n_minus+1), int(n_minus + 1 + n_max)):
        j_nu = j_nu + gamma_integration_result(x)
        #print j_nu
    
    j_nu = j_nu + n_integration(n_max, n_minus)
    return j_nu
    

n_summation(n_max)

#print gamma_integration_result(10)
```

```python
#approximate fit to thermal MBS emissivity eq. 72
#X = nu/nu_s
def eq_72_jnu(X):
    prefactor = (n_e * np.sqrt(2.) * np.pi * e**2. * nu_s)/(3. * special.kn(2., 1./theta_e) * c)
    f_of_X = (X**(1./2.) + 2**(11./12.) * X**(1./6.))**2. * np.exp(-X**(1./3.))
    ans = prefactor * f_of_X
    return ans

X = np.arange(0, 10000., 1.)
data = np.loadtxt('/home/alex/Documents/total_jnu_values3.txt')
pl.xlabel("$\\log_{10}(\\nu$/$\\nu_c$)")
pl.ylabel("log(j$_\\nu$)")

#since we are plotting nu/nu_c, and X = nu/nu_s, we
#need to plot (X * nu_s)/nu_c, eq_72(X)
#used sums up until nu/nu_c = 10000 (including missing point 5000)
pl.loglog(data[:, 0], data[:, 1], label='calculated j_nu')
pl.loglog((X * nu_s)/nu_c, eq_72_jnu(X), label='fitting function')
pl.legend()
pl.show()
```

```python
#we can see that the calculation is accurate up to nu/nu_c = 10^4

errordata = np.loadtxt('/home/alex/Documents/errordata.txt')
nu_ratio = errordata[:, 0]
calculated_jnu = errordata[:, 1]

fitting_function_jnu = eq_72_jnu(nu_ratio * nu_c / nu_s)

error = fitting_function_jnu/calculated_jnu - 1.

pl.xlabel("$\\log_{10}(\\nu$/$\\nu_c$)")
pl.ylabel("j$_\\nu eq.72$/j$_\\nu$ - 1")
pl.semilogx(nu_ratio, error)
```
