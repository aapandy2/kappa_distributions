```python
#THERMAL DISTRIBUTION
#plotting observer angle vs emissivity j
#for some reason, it fits exactly when n_e in our code is 1, rather than 1e5

data2 = np.loadtxt('/home/alex/Documents/thermal_jnu_vs_angle1.txt')


pl.xlabel("$\\theta$ (degrees)")
pl.ylabel("j$_\\nu$")
pl.plot(data2[:, 0], data2[:, 1], label='our calculation')

pl.plot(data[:, 0], data[:, 2], label='Eq. 72')
pl.plot(data[:, 0], data[:, 1], label='harmony')
pl.title('Thermal j$_\\nu$: harmony vs Our Calculation')
pl.legend(loc='lower right')
pl.show()
```

```python
#NONTHERMAL j_nu: iharmony vs our calculation 
#for gamma_min = 1, gamma_max = 1000, p = 3
n_e = 1.
B = 100.

data_nonthermal_iharmony = np.loadtxt('/home/alex/iharmony-mod/datafile_nonthermal.txt')
data_nonthermal_ours = np.loadtxt('/home/alex/Documents/jnu_power_law_ours.txt')

pl.xlabel("$\\log_{10}(\\nu$/$\\nu_c$)")
pl.ylabel("$\\log_{10}(j_\\nu / (n_eB)$)")
pl.loglog(data_nonthermal_ours[:, 0], (1./(n_e * B)) * data_nonthermal_ours[:, 1], label='our calculation')
pl.loglog(data_nonthermal_iharmony[:, 0], (1./(n_e * B)) * data_nonthermal_iharmony[:, 2], label='Eq. 58')
pl.loglog(data_nonthermal_iharmony[:, 0], (1./(n_e * B)) * data_nonthermal_iharmony[:, 1], label='harmony')
pl.title('Power Law j$_\\nu$: harmony vs Our Calculation')
pl.legend()
```
