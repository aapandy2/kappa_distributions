```python
import numpy as np
import pylab as pl
from scipy.interpolate import splrep, splev
%matplotlib inline
```

```python
x = np.arange(-np.pi, np.pi, 0.5)
y = np.sin(x)
pl.plot(x, y, 'o-')
pl.show()
```

```python
tck = splrep(x, y)
```

```python
pl.plot(x, y, 'o-')

x_fine_grid = np.arange(-np.pi, np.pi, 0.01)
y_fine_grid = splev(x_fine_grid, tck, der=1)
pl.plot(x_fine_grid, y_fine_grid)
```
