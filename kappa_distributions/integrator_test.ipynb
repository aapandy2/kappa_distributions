```python
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
theta = np.pi/3.
B = 30.
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)


#parameters for the power law distribution
p = 3
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.
```

```python
#power-law emissivity works out to 3e10

nu_ratio = np.arange(1., 1.5e11, 1.5e11/10.)

def eq_58(nu_ratio):
    prefactor = n_e_NT*(e**2.*nu_c)/c*(3.**(p/2.)*(p-1.)*np.sin(theta))/(2.*(p+1.)*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3. * p - 1.)/12.) * special.gamma((3. * p + 19.)/12.)
    ans = prefactor * gamma_fns * (nu_ratio/(np.sin(theta)))**(-(p - 1.)/2.)
    return ans

data_old = np.loadtxt('/home/alex/Documents/verification/pl_no_cutoff.txt')
x_old = data_old[:, 0]
y_old = data_old[:, 1]

data_new = np.loadtxt('/home/alex/pl_test.txt')
x_new = data_new[:, 0]
y_new = data_new[:, 1]


#pl.loglog(x_new, y_new, label='improved c code')
#pl.loglog(x_old, y_old, label='previous c code')
#pl.loglog(nu_ratio, eq_58(nu_ratio), label='fitting formula')
pl.loglog(x_old, abs(y_old/eq_58(x_old)-1.), label='old c code')
pl.loglog(x_new, abs(y_new/eq_58(x_new)-1.), label='new c code')
pl.title('Integrator Test Error')
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("data/fitting - 1")
pl.legend(loc='lower left')
pl.xlim([1., 1.5e11])
```

```python
#power-law absorptivity seems to work out to 3e10
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
theta = np.pi/3.
B = 30.
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)


#parameters for the power law distribution
p = 3
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.


data = np.loadtxt('/home/alex/Documents/pl_abs.txt')
x = data[:, 0]
y = data[:, 1]

nuratio = np.arange(1., 1e6, 1)

def eq_59(nuratio):
    nu = nuratio * nu_c
    term1 = n_e_NT * e**2./(nu*m*c) * ((3.**((p+1)/2.) * (p-1.))/(4. * gamma_min**(1.-p) - gamma_max**(1.-p)))
    term2 = special.gamma((3.*p + 2.)/12.) * special.gamma((3. * p + 22.)/12.)
    term3 = (nu/(nu_c * np.sin(theta)))**(-(p+2.)/2.)
    return term1*term2*term3

#print eq_59(100)
pl.loglog(x, eq_59(x), label='fit')
pl.loglog(x, y, label='data')
#pl.loglog(x, abs(eq_59(x)/y-1.))

pl.title('Power-Law Absorptivity')
pl.xlabel("$\\nu$/$\\nu_c$")
#pl.ylabel("data/fitting - 1")
pl.ylabel("j$_\\nu$")

pl.legend(loc='lower left')
```

```python
#kappa absorptivity for kappa = 3.5

data = np.loadtxt('/home/alex/kappa_abs.txt')
x = data[:, 0]
y = data[:, 1]

kappa = 3.5
n_e = 1.
w = 10.

def anu_low(nuratio):   
    nu = nuratio * nu_c
    term1 = 3. * 3.**(1./6.) * e * (-2. + kappa) * (kappa - 1.) * kappa
    term2 = n_e**2. * (2. * np.pi)**(1./3.) * (kappa * w)**(-2. + kappa)
    term3 = (1./np.sin(theta)) * special.gamma(5./3.)
    term4 = special.hyp2f1(-1./3. + kappa, 1. + kappa, 2./3. + kappa, -kappa * w)
    term5 = B*(-1.+3.*kappa)*((m*nu)/(B*e*np.sin(theta)))**(5./3.)
    f = term1*term2*term3*term4/term5
    return f
    
def anu_high(nuratio):
    nu = nuratio * nu_c
    term1 = 1./(3. * B * (2. + kappa))
    term2 = 2.**((-1./2.) - kappa) * e * (- 2. + kappa ) * (-1. + kappa) * kappa
    term3 = n_e**2 * np.pi**(1. - kappa) * (kappa * w)**(-2. + kappa)
    term4 = 1./np.sin(theta)
    term5 = 2.**(1. + kappa/2.) * np.pi**(kappa/2.) * special.gamma(2. + kappa/2.)
    term6 = 2**(1. + kappa/2.) * np.pi**(kappa/2.) + kappa * (2. * np.pi)**(kappa/2.)
    term7 = special.gammainc(1. + kappa/2., 2. * c * m * nu * np.pi / (B * e * np.sin(theta)))
    term8 = (B * e * np.sin(theta)/(c * m * nu))**((3. + kappa)/2.)
    f = term1 * term2 * term3 * term4 * (term5 - term6 * term7) * term8
    return f

def total_fitting_anu(x):
    term1 =  (1./(np.pi**4.))*(m*c)*anu_low(x)
    term2 = anu_high(x)
    aa = -7.7516
    ab = 3.64007
    ac = 0.692185
    a = (aa + ab*kappa)**(-ac)
    f = (1./((1./term1)**(a) + (1./term2)**(a)))**(1./a)
    return f

#pl.loglog(x, total_fitting_anu(x), label='fit')
#pl.loglog(x, y, label='data')
pl.loglog(x, abs(total_fitting_anu(x)/y-1.))
pl.legend()

pl.title('$\kappa = 3.5$ Absorptivity Error')
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("data/fitting - 1")
#pl.ylabel("j$_\\nu$")
```
