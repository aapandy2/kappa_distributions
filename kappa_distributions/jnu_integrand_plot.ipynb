```python
#want to plot the integrand in Eq. 60 of the paper
#with filled in values of I(...), delta(yn), 

#m = 9.1093826e-28
#c = 2.99792458e10
m = 1.
c = 1.
theta_e = 10.
#e = 4.80320680e-10
e = 1.
h = 6.6260693e-27
B = 30.
n_e = 1.
nu_c = (e * B)/(2 * np.pi * m * c)
p = 3
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.
theta = np.pi/3.
#nuratio = np.arange(0., 1.e11, 100000)
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)
gamma = np.arange(1., 50., 49./100.)
cos_xi = np.arange(-1., 1., 2./100.)


#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1.e1 * nu_c
theta = np.pi/3.
epsilon = 0.1
n_e = 1
#xi = np.arange(-np.pi/2, np.pi/2, (np.pi)/100)
#gamma = np.arange(1, 11, 0.1)
beta = np.sqrt(1. - 1./gamma**2.)
    
def delta_function(n, gamma, cos_xi):
    argument = (n * nu_c)/gamma - nu * (1 - beta * cos_xi * np.cos(theta))
    f = (1/(epsilon * np.sqrt(np.pi))) * np.exp(-argument**2./epsilon**2.)
    return f

def K_s(n, gamma, cos_xi):
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1.- cos_xi**2.)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1.- cos_xi**2.))/nu_c
    K_xx = M**2. * (special.jv(n, z))**2.
    K_yy = N**2. * (special.jvp(n, z, 1))**2.
    f = K_xx + K_yy
    #f = K_yy
    return f

def power_law_f(gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    #d is the differential dn_e/dgammadphid(cos(xi))
    d = (n_e_NT * (p - 1.) * gamma**(-p)) / (4 * np.pi * (gamma_min**(1. - p) - gamma_max**(1. - p)))
    #exponential cutoff term for power law distribution
    #cutoff = np.exp(-gamma/gamma_max)
    f = 1./(m**3. * c**3. * gamma**2. * beta) * d
    return f
     
def jnu_integrand(n, gamma, cos_xi):
    prefactor = (np.pi * e**2. * nu**2. * n_e)/(c * theta_e * special.kn(2., 1./theta_e))
    MJ_f = gamma * np.sqrt(gamma**2. - 1.) * np.exp(-gamma/theta_e)
    f = prefactor * power_law_f(gamma) * delta_function(n, gamma, cos_xi) * K_s(n, gamma, cos_xi)
    return f

#print jnu_integrand(1, 100, 0.5)
```

```python
import pylab as pl

gamma_mesh, xi_mesh = np.meshgrid(gamma, cos_xi)

n = 5

Z = jnu_integrand(n, gamma_mesh, xi_mesh)

pl.figure(figsize=(12, 8))
pl.contourf(gamma_mesh, xi_mesh, Z, 100)
pl.xlabel('gamma')
pl.ylabel('cos(xi)')
pl.title('n = '+str(n))
```

```python
gamma_mesh, xi_mesh = np.meshgrid(gamma, cos_xi)

n = 10
total_integrand = 0
for i in xrange(n):
    total_integrand = total_integrand + jnu_integrand(i, gamma_mesh, xi_mesh)

#pl.figure(figsize=(12, 8))
pl.contourf(gamma_mesh, xi_mesh, total_integrand, 100)
pl.xlabel('$\\gamma$')
pl.ylabel('$\\cos(\\xi)$')
pl.colorbar()
#pl.title('n = '+str(n))
```
