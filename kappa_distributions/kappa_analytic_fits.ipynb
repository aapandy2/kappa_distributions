```python
#parameters
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
B = 30.
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1. * nu_c
theta = np.pi / 3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

kappa = 6.5
w = 10.
nu_w = (w*kappa)**2. * nu_c * np.sin(theta)
nuratio = np.arange(1., 3.981e6, 100)
```

```python
kappa_data = np.loadtxt('/home/alex/Documents/kappa_fits/kappa_6_5.txt')
x = kappa_data[:, 0]
y = kappa_data[:, 1]

def kappa_low(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_w
    norm = 4. * np.pi * special.gamma(kappa-4./3.) / (3.**(7./3.) * special.gamma(kappa-2.))
    dimensional = (n_e * e**2.)/(c * kappa**2. * w**2.)
    ans = norm * dimensional * nu_w * X**(1./3.)
    return ans

def kappa_high(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_w
    term1 = (1./4.) * 3.**((kappa-1.)/2.) * (kappa-2.) * (kappa-1.)
    term2 = special.gamma(kappa/4. - 1./3.) * special.gamma(kappa/4. + 4./3.) 
    term3 = (n_e * e**2. * nu_c * np.sin(theta)) / c
    ans = term1*term2*term3 * X**(-(kappa-2.)/2.)
    return ans

def z_fit(nuratio, x):
    ans = (1./(1./kappa_low(nuratio)**x + 1./kappa_high(nuratio)**x))**(1./x)
    return ans

def test(nuratio):
    nu = nuratio * nu_c
    term1 = 1.35785*(c/(e**2. * n_e * np.sin(theta) * nu_c * (nu/(w**2. * nu_c * np.sin(theta)))**(1./3.)))**(0.458162)
    term2 = 0.176631*((c * (nu/(w**2. * nu_c * np.sin(theta)))**(0.75))/(e**2. * n_e * nu_c * np.sin(theta)))**(0.458162)
    ans = (1./(term1 + term2))**(2.18263)
    return ans

z_coeff = curve_fit(z_fit, x, y)[0]
print z_coeff
print 3 * kappa**(-3./2.)
```

```python
pl.loglog(x, y, label='data')
pl.loglog(x, z_fit(x, 3. * kappa**(-3./2.)), label='test')
#pl.semilogx(x, test(x)/y - 1., label='test')
#pl.xlim([1, 3.1e6])
pl.xlabel("$\\log_{10}(\\nu$/$\\nu_c$)")
pl.ylabel("log(j$_\\nu$)")
pl.title('$\kappa$ Fitting Formulae')
pl.legend()
pl.show()

pl.semilogx(x, z_fit(x, 3. * kappa**(-3./2.))/y - 1., label='test')
pl.show()
```
