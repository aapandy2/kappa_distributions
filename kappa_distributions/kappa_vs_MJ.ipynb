```python
# Maxwell-Juettner and Kappa distributions
m = 1.
c = 1.
theta = 10.

p = np.arange(-10., 10., 0.01)

#normalized Maxwell-Juettner distribution
def maxwell_juettner(p, theta): 
    gamma = np.sqrt(1. + (p/(m*c))**2.)
    f = 1./(4. * np.pi * m**3. * c**3. * theta * special.kn(2., 1./theta)) * np.exp(-gamma/theta)
    return f

#unnormalized kappa distribution
def kappa_not_normed(p, theta, kappa):
    f = (1. + (np.sqrt(1. + p**2.) - 1.)/(kappa * theta**1.))**(-(kappa + 1.))
    return f

#kappa distribution with numerical normalization
def kappa_numerical_norm(p, theta, kappa):
    norm = quad(lambda p: 4. * np.pi * p**2 * kappa_not_normed(p, theta, kappa), 0, np.inf)
    f = (1./norm[0]) * kappa_not_normed(p, theta, kappa)
    return f

#kappa distribution with analytic normalization
def kappa_analytic_norm(p, theta, kappa):
    norm = (4. * np.pi)**-1. * 1./((8. * special.beta(3./2. , kappa - 2.)/(2. * kappa - 1.)) * 
            (3. * special.hyp2f1(kappa + 1., 5./2., kappa + 1./2., 1. - 2./(kappa * theta**1.)) 
            + (kappa - 2.) * special.hyp2f1(kappa + 1., 3./2., kappa + 1./2., 1. - 2./(kappa * theta**1.))))
    f = (norm) * kappa_not_normed(p, theta, kappa)
    return f


#define test functions to check normalization
def MJ_norm_check(p):
    f = 4. * np.pi * p**2. * maxwell_juettner(p, theta)
    return f
def kappa_analytic_norm_check(p):
    f = 4. * np.pi * p**2. * kappa_analytic_norm(p, theta, 100)
    return f
def kappa_numerical_norm_check(p):
    f = 4. * np.pi * p**2. * kappa_numerical_norm(p, theta, 100)
    return f


#testing that all three distributions are normalized
#test1 = quad(MJ_norm_check, 0, np.inf)
#print test1[0]
#test2 = quad(kappa_analytic_norm_check, 0, np.inf)
#print test2[0]
#test3 = quad(kappa_numerical_norm_check, 0, np.inf)
#print test3[0]


#we think theta is different for MJ vs kappa
#calculating theta for MJ to test (second moment of distribution)

def MJ_test_theta(p):
    f = (1./3.) * ((4. * np.pi * p**4.)/np.sqrt(1 + p**2.)) * maxwell_juettner(p, theta)
    return f

def kappa_test_theta(p):
    f = ((4. * np.pi * p**4.)/np.sqrt(1 + p**2.)) * kappa_numerical_norm(p, theta, 150)
    return f

#theta_from_def_MJ = quad(MJ_test_theta, 0, np.inf)

#print theta_from_def_MJ[0]
#print theta


theta_from_kappa = quad(kappa_test_theta, 0, np.inf)

print theta
print (1./3.) * theta_from_kappa[0]
```
