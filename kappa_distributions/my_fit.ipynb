```python
#parameters
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
B = 30.
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1. * nu_c
theta = np.pi / 3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)
w = 10.
kappa = 3.2
nu_w = (w*kappa)**2. * nu_c * np.sin(theta)
nuratio = np.arange(1., 1e9, 100)
```

```python
kappa = 150.

nu_w = (w*kappa)**2. * nu_c * np.sin(theta)
#kappa_data = np.loadtxt('/home/alex/Documents/kappa_fits/kappa_6.txt')
kappa_data = np.loadtxt('/home/alex/Documents/paper_plots/kappa_150.txt')
#kappa_data = np.loadtxt('/home/alex/kappa_3_5.txt')
x = kappa_data[:, 0]
y = kappa_data[:, 1]

def kappa_low(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_w
    norm = 4. * np.pi * special.gamma(kappa-4./3.) / (3.**(7./3.) * special.gamma(kappa-2.))
    #norm = C_fit(kappa, 2.43091789,  0.45470688, -3.11539714)
    #norm = C_fit(kappa, 2.43,  0.454, -3.12)
    dimensional = (n_e * e**2. * nu_c * np.sin(theta))/(c)
    ans = norm * dimensional * X**(1./3.)
    return ans

def kappa_high(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_w
    term1 = (1./4.) * 3.**((kappa-1.)/2.) * (kappa-2.) * (kappa-1.)
    term2 = special.gamma(kappa/4. - 1./3.) * special.gamma(kappa/4. + 4./3.) 
    D = term1 * term2
    #D = 10.**D_fit(kappa, 0.5141287 , -0.99470623)
    #D = 10.**D_fit(kappa, 0.5 , -1)
    term3 = (n_e * e**2. * nu_c * np.sin(theta)) / c
    ans = D*term3 * X**(-(kappa-2.)/2.)
    return ans

def z_fit(nuratio):
    x = 3. * kappa**(-3./2.)
    #ans = (1./(1./kappa_low(nuratio)**x + 1./kappa_high(nuratio)**x))**(1./x)
    ans = (kappa_low(nuratio)**(-x) + kappa_high(nuratio)**(-x))**(-1./x)
    return ans

def ap(nuratio):
    a = 3. * kappa**(-3./2.)
    nu = nuratio * nu_c
    X_kappa = X = nu/nu_w
    dimensional = n_e * e**2. * nu_c * np.sin(theta)/c
    #C = 2.43091789*kappa**(0.45470688)-3.11539714
    #C = 2.431*kappa**(0.455)-3.115
    #C = 0.44*kappa - 0.32
    C = 11./25. * kappa - 8./25.
    #D = 10.**(0.5141287*kappa-0.99470623)
    D = 10.**(0.5*kappa-1.)
    ans = dimensional * ((C * X_kappa**(1./3.))**(-a) + (D * X_kappa**(-kappa/2.+1.))**(-a))**(-1./a)
    return ans

def ap_reduced(nuratio):
    a = 3. * kappa**(-3./2.)
    nu = nuratio * nu_c
    X_kappa = X = nu/nu_w
    dimensional = n_e * e**2. * nu_c * np.sin(theta)/c
    C = 11./25. * kappa - 8./25.
    D = 10.**(0.5*kappa-1.)
    ans = dimensional * X_kappa**(1./3.) * (C**-a + D**-a * X_kappa**(-2. * kappa**(-3./2.) + 1.5* kappa**(-1./2.)))**(-1./a)
    return ans

def ap_reduced1(nuratio):
    a = 3. * kappa**(-3./2.)
    nu = nuratio * nu_c
    X_kappa = X = nu/nu_w
    dimensional = n_e * e**2. * nu_c * np.sin(theta)/c
    ans = dimensional * X_kappa**(1./3.) * ((11.*kappa-8.)/25.) * (1. + ((22.*kappa-16.)/(5.*10.**(kappa/2.)))**(3.*kappa**(-3./2.)) * X_kappa**(-2. * kappa**(-3./2.) + 1.5* kappa**(-1./2.)))**(-1./a)
    return ans


def fit(nuratio, x):
    #x = 3. * kappa**(-3./2.)
    #x = 1./(21./20. *kappa - 77./50.)
    nu = nuratio * nu_c
    X_kappa = nu/nu_w
    dim = (n_e * e**2. * nu_c * np.sin(theta))/(c)
    A = 4. * np.pi * special.gamma(kappa-4./3.) / (3.**(7./3.) * special.gamma(kappa-2.))
    B = (1./4.) * 3.**((kappa-1.)/2.) * (kappa-2.) * (kappa-1.) * special.gamma(kappa/4. - 1./3.) * special.gamma(kappa/4. + 4./3.)
    #ans = (kappa_low(nuratio)**(-x) + kappa_high(nuratio)**(-x))**(-1./x)
    ans = dim * A * X_kappa**(1./3.) * (1. + (B/A)**(-x)*(X_kappa**(-(kappa-2.)/2.) * X_kappa**(-1./3.))**(-x))**(-1./x)
    return ans

coeff = curve_fit(fit, x, y, p0=[0.005])
print coeff

#pl.loglog(x, y, label='data')
#pl.loglog(x, ap_reduced1(x))
#pl.loglog(x, z_fit(x))
#pl.loglog(x, kappa_low(x))
#pl.loglog(x, kappa_high(x))
#pl.loglog(x, fit(x, 0.00557532), label='fit')
#pl.loglog(x, z_fit(x, 3. * kappa**(-3./2.)))
pl.semilogx(x, abs(y/fit(x, 0.00557532)-1.))
#pl.plot(x, ap_reduced(x)/ap_reduced1(x))
pl.ylim([-1., 1.])
pl.legend()
```
