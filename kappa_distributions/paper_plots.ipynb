```python
#y_fit = data_harmony[:, 2]
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
h = 6.6260693e-27
B = 30.
n_e = 1.
nu_c = (e * B)/(2 * np.pi * m * c)
p = 3
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.
theta = np.pi/3.
nuratio = np.arange(0., 1.e11, 100000)
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)
```

```python
#MJ vs angle abs
data_us = np.loadtxt('/home/alex/Documents/paper_plots/us_angle_abs.txt')
x_us = data_us[:, 0]
y_us = data_us[:, 1]

data_harmony = np.loadtxt('/home/alex/Documents/paper_plots/iharmony_angle.txt')
x_h = data_harmony[:, 0]
y_h = data_harmony[:, 3]


def eq_72(angle):
    theta = angle / 180. * np.pi
    nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)
    nuratio = 1.e2
    term1 = n_e * (np.sqrt(2.) * np.pi * e**2. * nu_s)/(3. * special.kn(2., 1./theta_e) * c)
    nu = nuratio * nu_c
    X = nu/nu_s
    term2 = (X**(1./2.) + 2.**(11./12.) * X**(1./6.))**2. * np.exp(-X**(1./3.))
    return term1*term2

def thermal_abs(angle):
    #theta = angle / 180. * np.pi
    nuratio = 1.e2
    nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)
    nu = nuratio * nu_c
    p_func = (2. * h * nu**3. / c**2.) * (np.exp(h*nu/(theta_e * m * c**2.))-1.)**(-1.)
    ans = eq_72(angle) / p_func
    return ans

pl.semilogy(x_us, y_us, label='symphony')
pl.semilogy(x_h, y_h, label='harmony')
pl.semilogy(x_us, thermal_abs(x_us), label='$\\bar{\\alpha_\\nu}$')
pl.xlabel("$\\theta$ (degrees)")
pl.ylabel("$\\alpha_\\nu$")
pl.legend(loc='lower right')
pl.xlim([5., 85.])
pl.show()

#error plot
pl.semilogy(x_us, abs(thermal_abs(x_us)/y_us - 1.))
pl.xlabel("$\\theta$ (degrees)")
pl.ylabel("$|\\bar{\\alpha_\\nu} /\\alpha_\\nu - 1|$")
pl.xlim([5., 85.])
pl.show()

```

```python
#pl vs nu emiss
data_us = np.loadtxt('/home/alex/Documents/paper_plots/us_pl_nu_emiss.txt')
x_us = data_us[:, 0]
y_us = data_us[:, 1]

data_harmony = np.loadtxt('/home/alex/Documents/paper_plots/iharmony_pl_nu.txt')
x_h = data_harmony[:, 0]
y_h = data_harmony[:, 1]

def eq_58(nu_ratio):
    prefactor = n_e_NT*(e**2.*nu_c)/c*(3.**(p/2.)*(p-1.)*np.sin(theta))/(2.*(p+1.)*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3. * p - 1.)/12.) * special.gamma((3. * p + 19.)/12.)
    ans = prefactor * gamma_fns * (nu_ratio/(np.sin(theta)))**(-(p - 1.)/2.)
    return ans

pl.loglog(x_us, y_us, label='symphony')
pl.loglog(x_h, y_h, label='harmony')
pl.loglog(nuratio, eq_58(nuratio), label='$\\bar{j_\\nu}$')
pl.legend(loc='lower left')
pl.ylim([1e-33, 1e-21])
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("$j_\\nu$")
pl.show()

#error plot
pl.loglog(x_us, abs(eq_58(x_us)/y_us - 1.))
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("$|\\bar{j_\\nu} /j_\\nu - 1|$")
pl.show()
```

```python
#pl vs nu abs
data_us = np.loadtxt('/home/alex/Documents/paper_plots/us_pl_nu_abs.txt')
x_us = data_us[:, 0]
y_us = data_us[:, 1]

data_harmony = np.loadtxt('/home/alex/Documents/paper_plots/iharmony_pl_nu.txt')
x_h = data_harmony[:, 0]
y_h = data_harmony[:, 3]

#y_fit = data_harmony[:, 2]
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
B = 30.
n_e = 1.
nu_c = (e * B)/(2 * np.pi * m * c)
p = 3
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.
theta = np.pi/3.
nuratio = np.arange(0., 1.e11, 100000)

def eq_59(nu_ratio):
    nu = nu_ratio * nu_c
    prefactor = n_e_NT * (e**2.)/(nu*m*c) * (3.**((p+1.)/2.) * (p-1.))/(4.*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3.*p+2.)/12.) * special.gamma((3.*p+22.)/12.)
    ans = prefactor * gamma_fns * (nu_ratio/np.sin(theta))**(-(p+2.)/2.)
    return ans

pl.loglog(x_us, y_us, label='symphony')
pl.loglog(x_h, y_h, label='harmony')
pl.loglog(nuratio, eq_59(nuratio), label='$\\bar{\\alpha_\\nu}$')
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("$\\alpha_\\nu$")
pl.legend(loc='lower left')
pl.show()


#error plot
pl.loglog(x_us, abs(eq_59(x_us)/y_us - 1.))
pl.xlabel("$\\nu$/$\\nu_c$")
pl.ylabel("$|\\bar{\\alpha_\\nu} /\\alpha_\\nu - 1|$")
pl.show()
```
