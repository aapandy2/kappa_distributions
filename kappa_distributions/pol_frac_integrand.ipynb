```python
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
B = 30.
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)


#parameters for the power law distribution
p = 3
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1.e6 * nu_c
theta = np.pi/3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

#max number of terms in sum
n_max = 30

#eq. 67 gives location of peak for n integration for thermal; in
#the n integration we integrate from end of sum (n_max) to C * n_peak_67
C = 10.
#tolerance is the minimum allowed contribution (ans/tolerance); smaller contributions are ignored
tolerance = 1.e5
#n_peak_power_law = 50

def K_s(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1 - cos_xi**2.)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1 - cos_xi**2.))/nu_c
    K_xx = M**2. * (special.jv(n, z))**2.
    K_yy = N**2. * (special.jvp(n, z, 1))**2.
    #ans = K_xx + K_yy
    #print n
    ans = -2. * M * N * special.jv(n, z) * special.jvp(n, z, 1)
    return ans    

def norm_integrand(gamma):
    f = gamma**-p * np.exp(-gamma/gamma_max)
    return f

def norm_power_law():
    norm_prefactor = (n_e_NT*(p-1.)/(4.*np.pi*(gamma_min**(1.-p)-gamma_max**(1.-p))))
    norm_integral_result = 4.*np.pi*quad(norm_integrand, 1., np.inf)[0]
    norm = norm_prefactor * norm_integral_result
    return 1./norm

normalization = norm_power_law()
#print normalization

def fit(x):
    nu_factor = nu/nu_c / 1.e6
    ans = 1.33322780e-06 * x / nu_factor
    return ans

def power_law_f(gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    #d is the differential dn_e/dgammadphid(cos(xi))
    d = (n_e_NT * (p - 1.) * gamma**(-p)) / (4 * np.pi * (gamma_min**(1. - p) - gamma_max**(1. - p)))
    #exponential cutoff term for power law distribution
    cutoff = np.exp(-gamma/gamma_max)
    f = 1./(m**3. * c**3. * gamma**2. * beta) * d * cutoff
    return f
     
def I(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    #this involves f; this current version uses power_law_f; 2pi comes from integrating out dphi
    ans = (2. * np.pi * e**2. * nu**2.)/c * (m**3. * c**3. * gamma**2. * beta * 2 * np.pi) * power_law_f(gamma) * K_s(gamma, n)  
    return ans

def gamma_integrand(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c)/(gamma * nu * beta * np.cos(theta))
    ans = 1./(nu * beta * abs(np.cos(theta))) * I(gamma, n)
    return ans

def gamma_integration_result(n):
    gamma_minus = ((n*nu_c)/nu - abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    gamma_plus = ((n*nu_c)/nu + abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    
    def integrand(gamma):
        f = gamma_integrand(gamma, n)
        return f
    gamma_minus_new = fit(n) - (fit(n)-gamma_minus)/10.
    gamma_plus_new = fit(n) + (gamma_plus-fit(n))/10.
    #print gamma_minus_new
    #print gamma_plus_new
    
    #if(nu/nu_c <= 1.e3):
        #result = quad(integrand, gamma_minus_new, gamma_plus_new, epsabs = 0, epsrel = 1e-08)[0]
    #    result = quad(integrand, gamma_minus, fit(n), epsabs = 0, epsrel = 1e-08)[0]
    #    result = result + quad(integrand, fit(n), gamma_plus, epsabs = 0, epsrel = 1e-08)[0]
    #elif(nu/nu_c > 1.e3 and nu/nu_c < 3.e8):
    #    result = quad(integrand, gamma_minus_new, gamma_plus_new, epsabs = 0, epsrel = 1e-08)[0]
    #else:
    result = quad(integrand, gamma_minus, fit(n), epsabs = 0, epsrel = 1e-08)[0]
    if(np.isnan(result)):
        result = 0.
    return result


#integrate over n after evaluating (n_max) terms of sum
def n_integration(n_max, n_minus):
    if(n_max < n_minus):
        n_max = int(n_minus + 1.)
    
    n_start = int(n_max)
    ans = 0.
    contrib = 0.
    i = 0
    delta_n = 1e2
    deriv_tol = 1e-5
    while(abs(contrib) >= abs(ans/tolerance)):
        deriv = derivative(gamma_integration_result, n_start, dx = 1)
        if(abs(deriv) < deriv_tol):
            delta_n = 10 * delta_n
            #delta_n = 1. * delta_n
        contrib = quad(gamma_integration_result, n_start, n_start + delta_n, epsabs = 0, epsrel = 1e-08)[0]
        ans = ans + contrib
        n_start = n_start + delta_n        
        i = i + 1
        #print i
    print n_start + delta_n
    return ans

#evaluate (n_max) number of terms of the sum
def n_summation(n_max):
    j_nu = 0
    n_minus = (nu/nu_c) * abs(np.sin(theta))
        
    for x in range (int(n_minus+1), int(n_minus + 1 + n_max)):
        j_nu = j_nu + gamma_integration_result(x)
    
    j_nu = j_nu + n_integration(n_max, n_minus)
    return j_nu

#print nu / nu_c
#n_summation(n_max)
```
