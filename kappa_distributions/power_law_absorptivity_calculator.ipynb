```python
#constants in natural units
#m = 1.
#c = 1.
#theta_e = 10.
#nu_c = 1
#B = 10.
#e = 1

#constants in Gaussian cgs units (values from harmony)
m = 9.1093826e-28
c = 2.99792458e10
e = 4.80320680e-10
B = 30.0
#T_e = 10.
#k = 1.3806505e-16
theta_e = 10.
#n_e = 1.

#other parameters to be set
theta = np.pi/3.
epsilon = 0.01
n_e = 1
nu_c = e * B / (2. * np.pi * m * c)
nu = 1.e0 * nu_c
nu_s = (2./9.) * nu_c * theta_e**2 * np.sin(theta)
C = 10.

#parameters for the PL distribution
p = 3.
gamma_min = 1.
gamma_max = 1000.
n_e_NT = 1.
gamma_cutoff = 1000.
```

```python
nu = 1e0 * nu_c
n_peak = 10000
```

```python
#functions K_stokes, where the subscript is the Stokes mode
def K_s(n, gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c) / (gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1. - cos_xi**2)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1. - cos_xi**2))/nu_c
    K_xx = M**2 * (special.jv(n, z))**2
    K_yy = N**2 * (special.jvp(n, z, 1))**2
    ans = K_xx + K_yy
    return ans

def K_Q(n, gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c) / (gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1. - cos_xi**2)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1. - cos_xi**2))/nu_c
    K_xx = M**2 * (special.jv(n, z))**2
    K_yy = N**2 * (special.jvp(n, z, 1))**2
    ans = K_xx - K_yy
    return ans

def K_U(n, gamma):
    ans = 0.
    return ans

def K_V(n, gamma):
    beta = np.sqrt(1. - 1./gamma**2.)
    cos_xi = (gamma * nu - n * nu_c) / (gamma * nu * beta * np.cos(theta))
    M = (np.cos(theta) - beta * cos_xi)/np.sin(theta)
    N = beta * np.sqrt(1. - cos_xi**2)
    z = (nu * gamma * beta * np.sin(theta) * np.sqrt(1. - cos_xi**2))/nu_c
    ans = -2. * M * N * special.jv(n, z) * special.jvp(n, z, 1)
    return ans

#differential operator D in terms of gamma, eq. 42; this is specific to the PL distribution
def D(gamma):
    PL_norm = 1.00199514442
    prefactor = (np.pi * nu / (m * c**2)) * (n_e_NT * (p - 1.))/(gamma_min**(1.-p) - gamma_max**(1.-p))
    term1 = ((-p-1.)*np.exp(-gamma/gamma_cutoff)*gamma**(-p-2.))/(np.sqrt(gamma**2. - 1.))
    term2 = (np.exp(-gamma/gamma_cutoff) * gamma**(-p-1.))/(gamma_cutoff * np.sqrt(gamma**2. - 1.))
    term3 = (np.exp(-gamma/gamma_cutoff) * gamma**(-p))/(gamma**2. - 1.)**(3./2.)
    f = PL_norm * prefactor * (term1 - term2 - term3)
    return f

def absorptivity_integrand(gamma, n):
    beta = np.sqrt(1. - 1./gamma**2.)    
    prefactor = - c * e**2 / (2. * nu)
    cos_xi = (gamma * nu - n * nu_c) / (gamma * nu * beta * np.cos(theta))
    f = prefactor * gamma**2 * beta * D(gamma) * K_s(n, gamma) * 1./(nu * beta * abs(np.cos(theta)))
    return f

def gamma_integration_result(n):
    gamma_minus = ((n*nu_c)/nu - abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    gamma_plus = ((n*nu_c)/nu + abs(np.cos(theta)) * np.sqrt(((n*nu_c)/nu)**2. - (np.sin(theta))**2.))/(np.sin(theta))**2.
    
    def integrand(gamma):
        f = absorptivity_integrand(gamma, n)
        return f
    
    result = quad(integrand, gamma_minus, gamma_plus, epsabs = 0, epsrel = 1.e-8)[0]
    return result

#integrate over n after evaluating (n_max) terms of sum
def n_integration(n_max, n_minus):
    if(n_max < n_minus):
        n_max = n_minus
    ans = quad(gamma_integration_result, n_max, C * n_peak, epsabs = 1e-29)[0]
    return ans

#evaluate (n_max) number of terms of the sum
def n_summation(n_max):
    j_nu = 0
    n_minus = (nu/nu_c) * abs(np.sin(theta))
        
    for x in range (int(n_minus+1), int(n_minus + 1 + n_max)):
        j_nu = j_nu + gamma_integration_result(x)
        #print j_nu
    
    j_nu = j_nu + n_integration(n_max, n_minus)
    return j_nu
```
