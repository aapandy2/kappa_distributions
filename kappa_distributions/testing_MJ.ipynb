```python
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10
e = 4.80320680e-10
B = 30.
h = 6.6260693e-27
nu_c = (e * B)/(2 * np.pi * m * c)
#nuratio = np.arange(1., 1.e6, 1)

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
#nu = 5. * nu_c
theta = 60. * np.pi / 180.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)
```

```python
#Stokes Q
data = np.loadtxt('/home/alex/Documents/paper_plots/MJ_q_T_10.txt')
x = data[:, 0]
y = -data[:, 1]

def lowq(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = n_e * e**2. * nu_c / c
    ans = (2.**(7./3.)*np.pi/27.)*((17.*theta_e**(7./10.)-23.)/(34.*theta_e**(7./10.)-36.))*np.sin(theta)*X**(1./3.)
    return prefactor * ans

def highq(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = n_e * e**2. * nu_c / c
    ans = (2.**(1./2.)*np.pi/27.) * np.sin(theta) * X * np.exp(-X**(1./3.))
    return prefactor * ans

def totalq(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = n_e * e**2. * nu_c / c
    A = (7.*theta_e**(24./25.)+35.)/(10.*theta_e**(24./25.)+75.)
    ans = (2.**(1./2.)*np.pi/27.) * np.sin(theta) * (X**(1./2.) + A*2.**(11./12.)*X**(1./6.))**2. * np.exp(-X**(1./3.))
    return prefactor * ans


pl.loglog(x, totalq(x))
pl.loglog(x, y)
```

```python
#Stokes V
data = np.loadtxt('/home/alex/Documents/paper_plots/MJ_v_T_10.txt')
x = data[:, 0]
y = -data[:, 1]

def lowv(nuratio):
    prefactor = n_e * e**2. * nu_c / c
    ans = 1./theta_e * (9./25.-21./25. * np.sin(theta - 28./25.))
    return prefactor*ans + nuratio*0.

def highv(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = n_e * e**2. * nu_c / c
    ans = 1./theta_e * (7./50. - 2./3. * np.sin(theta - 1.367)) * X**(3./5.) * np.exp(-X**(1./3.))
    return prefactor*ans

def totalv(nuratio):
    nu = nuratio * nu_c
    X = nu/nu_s
    prefactor = n_e * e**2. * nu_c / c * np.exp(-X**(1./3.))
    ans = (37.-87.*np.sin(theta-28./25.))/(100.*(theta_e+1.)) * (1.+(theta_e**(3./5.) /25.+7./10.)*X**(9./25.))**(5./3.)
    return prefactor * ans

pl.loglog(x, totalv(x))
pl.loglog(x, y)
```
