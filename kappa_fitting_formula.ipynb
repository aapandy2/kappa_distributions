```python
#constants using iharmony's units
m = 9.1093826e-28
c = 2.99792458e10
theta_e = 10.
e = 4.80320680e-10
B = 30.
#n_e = 1.e5
nu_c = (e * B)/(2 * np.pi * m * c)


#parameters for the kappa distribution
kappa = 3.5
gamma_cutoff = 1000.

#variables to be set (frequency, observer angle, width of delta function, electron number density, 
# sum index n)
nu = 1e0 * nu_c
theta = np.pi/3.
epsilon = 0.01
n_e = 1.
nu_s = (2./9.) * nu_c * theta_e**2. * np.sin(theta)

#power law parameters
n_e_NT = 1.
gamma_min = 1.
gamma_max = 1000.
p = 4.5
theta_e = 10
```

```python
def eq_72_jnu(nuratio):
    X = nuratio * nu_c 
    prefactor = (n_e * np.sqrt(2.) * np.pi * e**2. * nu_s)/(3. * special.kn(2., 1./theta_e) * c)
    f_of_X = (X**(1./2.) + 2**(11./12.) * X**(1./6.))**2. * np.exp(-X**(1./3.))
    ans = prefactor * f_of_X
    return ans

def eq_55(nuratio):
    X = nuratio * nu_c / nu_s
    j_nu = (2.**(4./3.) * np.pi * n_e * e**2. * nu_s * X**(1./3.))/(3. * c * theta_e**2.)
    return j_nu

def eq_58(nuratio):
    prefactor = n_e_NT*(e**2.*nu_c)/c*(3.**(p/2.)*(p-1.)*np.sin(theta))/(2.*(p+1.)*(gamma_min**(1.-p)-gamma_max**(1.-p)))
    gamma_fns = special.gamma((3. * p - 1.)/12.) * special.gamma((3. * p + 19.)/12.)
    ans = prefactor * gamma_fns * (nuratio/(np.sin(theta)))**(-(p - 1.)/2.)
    return ans

def eq_57(nuratio):
    X = nuratio * nu_c / nu_s
    j_nu = (n_e * np.sqrt(2.) * np.pi * e**2. * nu_s) / (6. * theta_e**2. * c) * X * np.exp(-(X**(1./3.)))
    return j_nu
```
